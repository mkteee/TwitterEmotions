```python
import pandas as pd
```

```python
# Ignore the copy names again, this refers to the original dataset after cleaning the URL and the quotation marks
df = pd.read_csv("train_cleanFull3.csv", sep='\t', header=None, names=['Emotion', 'Text'])
```

```python
df = pd.DataFrame(df)
```

```python
# Count the spaces which means the words - 1 for visual help
print(df['Text'].str.count(' ') + 1)
```

```python
# Add a column with the values
df['Words'] = df['Text'].str.count(' ') + 1
```

```python
# Check the longest tweets. Left is amount of words, right is accumulated count.
df.groupby(['Words']).size()
```

```python
grouped_df = df.groupby(['Words']).size()
```

```python
# Show more stuff
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Used this, so I could use it for diagrams in Excel
grouped_df.to_csv("wordLengthCleaned.csv")
```

```python
# This was for manually cleaning up the dataset.
# As you can see it will just print "Done!" now, because no entry will be longer than 60 Words.
# Before it printed "i", which was the index and "sent", which was the sentence for that index.
# As stated in the report, most tweets looked like this: "@USERNAME @USERNAME @USERNAME @USERNAME @USERNAME [...] and here was an actual tweet"
# I manually shortened that data through the row index printed, because I couldn't be 100% sure that the data was always in this shape.
for i, sent in enumerate(df["Text"]):
    if (df["Words"][i]) > 60:
        print(i)
        print(sent)
print("Done!")
```
