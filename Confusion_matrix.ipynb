### Showing Confusion-Matrix with seaborn

This is just an example of how I manually created the confusion matrix for my report.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_matrix = [[2497, 211, 325, 356, 260, 185],
              [361, 2512, 168, 353, 330, 110],
              [265, 82, 2976, 166, 195, 152],
              [364, 213, 159, 2437, 126, 176],
              [333, 265, 282, 160, 2617, 178],
              [204, 80, 145, 219, 198, 3348]]
```

```python
all_matrix
```

```python
# Wir heben den Background etwas ab und vergrößern die Fonts.
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'}, font_scale=2)

# Wir stellen das Ergebnis der Prediction auf dem gesamten Datensatz dar. 
df_cm = pd.DataFrame(all_matrix)
plt.figure(figsize = (12,8))
sns.heatmap(df_cm, annot=True, fmt='g', xticklabels=['Anger', 'Disgust', 'Fear', 'Sad', 'Surprise', 'Joy'], 
            yticklabels=['Anger', 'Disgust', 'Fear', 'Sad', 'Surprise', 'Joy'])

plt.xlabel('Predicted Label', fontsize = 25) 
plt.ylabel('True Label', fontsize = 25) 
```
